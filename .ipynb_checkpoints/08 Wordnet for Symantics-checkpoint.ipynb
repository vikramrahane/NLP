## read a word from user and print all meanings of word      (its own name not be their)
## read a word from user and print all noun meanings of it 

```python
word=input("Enter a word: ")
synset=set()
for w in wordnet.synsets(word):        # all meanings except its own
    if w.lemmas()[0].name()!=word:
        synset.add(w.lemmas()[0].name())
synset
```

```python
word=input("Enter a word: ")
print("Meanings of ",word,' are: ')
synset=set()
for w in wordnet.synsets(word):
    if '.n.' in w.name() and w.lemmas()[0].name()!=word:    # select only 'Noun' and same word not taken
        print(w.name(),w.definition())
        print("-------------------------")
synset
```

```python
word=input("Enter a word: ")
print("Meanings of ",word,' are: ')

for w in wordnet.synsets(word,pos='n'):   # select only 'Noun'
    print(w.name(),"  ",w.definition())
    print("-------------------------")
```

# Find the synonyms of all the adjective in the sentence

```python
sent='''The fiber is most often spun into yarn or thread
    and used to make a soft, rigid, and flexible textile.'''
```

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
adj=[x for x in pos_tag(word_tokenize(sent)) if x[1].startswith('JJ')]
```

```python
adj
```

```python
for adjective in adj:
    print("Meanings of ",adjective[0]," are: ")
    for w in wordnet.synsets(adjective[0]):        # all meanings except its own
        if w.lemmas()[0].name()!=adjective[0]:
            print(w.lemmas()[0].name())
    print("--------------")
```

```python
## 2nd Logic Better One
adj=[word[0] for word in pos_tag(word_tokenize(sent)) 
    if word[1].startswith('JJ')]
```

```python
adj
```

```python
for word in adj:
    syns=set()
    for syn in wordnet.synsets(word):
        syns.add(syn.lemmas()[0].name())
    syns.discard(word)  # remove 
    print("Synonyms of ",word," are ", syns)
```

## Antonyms or opposite words

```python
ant=set()
for synset in wordnet.synsets("full"):
    for lem in synset.lemmas():
        if lem.antonyms():  # if antonyns are available then add
            ant.add(lem.antonyms()[0].name())
print("Antonyms: ",ant)
```
