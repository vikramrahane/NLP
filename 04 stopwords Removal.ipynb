## Remove Punctuations

```python
text='''Sachin Ramesh Tendulkar, born 24 April 1973) is an Indian former international cricketer who captained 
the Indian national team. He is widely regarded as one of the greatest batsmen in the history of cricket.[4] 
Hailed as the world's most prolific batsman of all time, he is the all-time highest run-scorer in both ODI and
Test cricket with more than 18,000 runs and 15,000 runs, respectively.[5] He also holds the record for receiving
the most player of the match awards in international cricket.[6] Tendulkar was a Member of Parliament, 
Rajya Sabha by nomination from 2012 to 2018.'''
```

```python
text=text.lower()
text
```

```python
import string
string.punctuation
```

```python
from nltk.tokenize import word_tokenize
tkn=word_tokenize(text)
tkn
```

```python
clean=[]
for i in tkn:
    if i not in string.punctuation:
        clean.append(i)
print(clean)
```

```python
clean
```

```python
" ".join(clean)
```

## remove stopwords

```python
from nltk.corpus import stopwords
```

```python
swords=stopwords.words('english')
print(swords)
```

```python
no_sword=[]
for i in clean:
    if i not in swords:
        no_sword.append(i)
print(no_sword)
```

```python
' '.join(no_sword) 
```

#### frequency plot

```python
from nltk import FreqDist
```

```python
freq= FreqDist(word_tokenize(text))
freq.plot(20, cumulative=False)
```

```python
freq= FreqDist(clean)
freq.plot(20, cumulative=False)
```

```python
freq= FreqDist(no_sword)
freq.plot(20, cumulative=False)
```

```python
## Function to clean(remove punctuatuion and stop words) text and plot frequency
def clean_text(text):
    tokens1= word_tokenize(text)
    tokens2=[word.lower() for word in tokens1
            if word not in string.punctuation]
    tokens3= [word for word in tokens2
             if word not in swords]
    return tokens3
def clean_text_plot(text):
    tokens1= word_tokenize(text)
    tokens2=[word.lower() for word in tokens1
            if word not in string.punctuation]
    tokens3= [word for word in tokens2
             if word not in swords]
    freq=FreqDist(tokens3)
    freq.plot(20)
```

```python
len(clean_text(text))
```

```python
print(clean_text(text))
```

```python
clean_text_plot(text)
```
