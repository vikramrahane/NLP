```python
# Dataset bbc-news-data.txt
# Locations : https://mitu/.co.in/dataset

import matplotlib.pyplot as plt
import pandas as pd
```

```python
f1 = open("../NLP/bbc-text-data.csv")
data = f1.readlines()
```

```python
len(data)
```

```python
# Feature Extraction 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
xvect = tfidf.fit_transform(data)
xvect.shape
```

```python
from sklearn.cluster import KMeans
```

```python
sse = []
for k in range(1,40):
    km = KMeans(n_clusters=k, random_state=23)
    km.fit_predict(xvect)
    sse.append(km.inertia_)
```

```python
plt.plot(range(1, 40), sse)
```
