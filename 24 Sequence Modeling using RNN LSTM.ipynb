## Prepare the Data

```python
alphabates='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
len(alphabates)
```

## PreProcess the data

```python
for i,v in enumerate(alphabates):
    print(i,v)
```

```python
int_to_char=dict(enumerate(alphabates))
int_to_char
```

```python
char_to_int=dict((v,i) for i,v in enumerate(alphabates))
char_to_int
```

```python
int_to_char[5]
```

```python
char_to_int['P']
```

```python
seq_length=1
```

```python
x=[]
y=[]
for i in range(len(alphabates)-seq_length):
    seq_in=[alphabates[i]]
    seq_out=alphabates[i+1]
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
    print(seq_in,'->',seq_out)
```

```python
x
```

```python
y
```

### One hot encoding

```python
from keras.utils import to_categorical
```

```python
y_new=to_categorical(y)
y_new
```

```python
y_new.shape
```

```python
import numpy as np
```

```python
np.array(x).shape
```

```python
x=np.reshape(x,(25,1,1))
```

```python
x.shape
```

```python
x
```

```python
# to scaling if min value is zero in data then divide by max value it normalize data
```

```python
x=x/25  # divide by 25 because 'z' is in output but not in input, so the maximum scaling is less than 1 
        # If the scaled data maximum is 1 then in output 'z' can't be predicted 
```

```python
x
```

### Build the Model

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
model=Sequential()

model.add(Dense(25,input_shape=(1,1),activation='relu'))
model.add(LSTM(128))
model.add(Dense(26,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y_new,epochs=1000,batch_size=1)
```

```python
model.evaluate(x,y_new,batch_size=1)
```
