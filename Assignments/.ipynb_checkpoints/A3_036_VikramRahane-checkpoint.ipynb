# Q1

### read a word from user and print all meanings of word (its own name not be their)

```python
from nltk.corpus import wordnet
```

```python
word=input("Enter a word: ")
synset=set()
for w in wordnet.synsets(word):        # all meanings except its own
    if w.lemmas()[0].name()!=word:
        synset.add(w.lemmas()[0].name())
synset
```

### read a word from user and print all noun meanings of it

```python
word=input("Enter a word: ")
print("Meanings of ",word,' are: ')

for w in wordnet.synsets(word,pos='n'):  
    if w.lemmas()[0].name()!=word:
        print(w.name(),"  ",w.definition())
        print("-------------------------")
```

### 3 stemmer

```python
from nltk.stem import PorterStemmer,LancasterStemmer
```

```python
ps=PorterStemmer()
ls=LancasterStemmer()
```

```python
verbs=[]
for word, tag in pos_tag(word_tokenize(maharashtra)):    
    if tag.startswith('VB') and word.isalpha():
        verbs.append(word)
verbs
```

```python
# PorterStemmer
for word in verbs:
    print(ps.stem(word))
```

```python
# LancasterStemmer
for word in verbs:
    print(ls.stem(word))
```

### 4 lemmatize

```python
past_tense_verbs=set()
for word, tag in pos_tag(word_tokenize(maharashtra)):    
    if tag.startswith('VBD') and word.isalpha():
        past_tense_verbs.add(word)
past_tense_verbs=list(past_tense_verbs)
past_tense_verbs
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wnl= WordNetLemmatizer()
for word in past_tense_verbs:
    print(wnl.lemmatize(word,pos='v'))
```

### 5 synonyms of adjectives

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
adj=set() 
for word in pos_tag(word_tokenize(maharashtra)): 
    if (word[1].startswith('JJ') and word[0].isalpha()):
        adj.add(word[0])
adj=list(adj)
print(adj)
```

```python
for word in adj:
    syns=set()
    for syn in wordnet.synsets(word):
        syns.add(syn.lemmas()[0].name())
    syns.discard(word)  # remove 
    print("Synonyms of ",word," are ", syns)
```

### 6 antonyms of verbs

```python
verbs=set()
for word, tag in pos_tag(word_tokenize(maharashtra)):    
    if tag.startswith('VB') and word.isalpha():
        verbs.add(word)
verbs=list(verbs)  # List of unique verbs
print(verbs)
```

```python
ant=set()
for verb in verbs:
    for synset in wordnet.synsets(verb):
        for lem in synset.lemmas():
            if lem.antonyms():  # if antonyns are available then add
                ant.add(lem.antonyms()[0].name())
    print("Antonyms of ",verb,": ",ant)
    print()
```
