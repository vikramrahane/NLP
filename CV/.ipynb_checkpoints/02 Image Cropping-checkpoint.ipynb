```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
img = cv2.imread('caves.jpg')
```

```python
plt.imshow(img);
```

```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img);
```

## Relative Cropping

```python
height, width, _ = img.shape
```

```python
height
```

```python
width
```

```python
startRow = int(height*.15)
startCol = int(width*.85)
endRow = int(height*.85)
endCol = int(width*.15)
```

```python
new_img = rgb_img[startRow:endRow]
```

```python
plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.grid()
plt.title("ORIGINAL IMAGE")
plt.imshow(rgb_img)

plt.subplot(1,2,2)
plt.title("CROPPED IMAGE")
plt.imshow(new_img)
```
