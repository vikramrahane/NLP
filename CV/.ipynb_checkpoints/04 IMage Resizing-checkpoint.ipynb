```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
img= cv2.imread('dog.jpg')
```

```python
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
newImg= cv2.resize(img,(0,0), fx=0.75, fy=0.75)
```

```python
newImg.shape
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Resized Image')
plt.imshow(newImg);
```

```python
newImg= cv2.resize(img,(250,150))
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Resized Image')
plt.imshow(newImg);
```

```python
newImg= cv2.resize(img,(150,200))
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Resized Image')
plt.imshow(newImg);
```
